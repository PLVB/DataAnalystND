```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('Sales Target.csv')
df1=df.groupby(['Category'])['Sales Target'].sum()
ax=df1.plot()
ax.set_ylabel('Sales Target')
ax.set_xlabel('Category')
ax.set_title('Sales Target by Category')
plt.show()

df2=df.groupby(['Segment'])['Sales Target'].sum()
ax=df2.plot()
ax.set_ylabel('Sales Target')
ax.set_xlabel('Segment')
ax.set_title('Sales Target by Segment')
plt.show()

df2.plot.pie(y=df2.index,
           shadow=False,
           startangle=90,
           autopct='%1.1f%%')
plt.title('Sales Target by Segment')
plt.axis('equal')
plt.tight_layout()
plt.show()

df1.plot.pie(y=df1.index,
           shadow=False,
           startangle=90,
           autopct='%1.1f%%')
plt.title('Sales Target by Category')
plt.axis('equal')
plt.tight_layout()
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('Sales Target.csv')

ax=df.groupby(['Category', 'Segment'])['Sales Target'].sum().unstack().plot(kind='bar', stacked=True)
ax.set_ylabel('Sales Target')
ax.set_xlabel('Category')
ax.set_title('Sales Target by Category&Segment')
ax.legend(loc='upper right', bbox_to_anchor=(1.5, 0.5))
plt.show()
```
