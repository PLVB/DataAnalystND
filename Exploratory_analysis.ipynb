```python
#Univariate with Horizental Bar
#Horizental bar chart - Bar charts are arguably the simplest data visualization. They map categories to numbers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
trips = pd.read_csv("2017-fordgobike-tripdata.csv", index_col=0)
Male = mpatches.Patch(color='blue', label='Male')
Female = mpatches.Patch(color='orange', label='Female')
Others = mpatches.Patch(color='green', label='Others')

ax=(trips['member_gender']).value_counts().plot.barh()
ax.set_ylabel('Gender', fontsize=13, color='blue')
ax.set_xlabel('Trip Counts', fontsize=13, color='blue')
ax.set_title('Fordgobike trip count by gender',  fontsize=13, color='blue')
total=trips['member_gender'].value_counts()
plt.legend(handles=[Male,Female,Others], loc=0)
```

```python
#Horizental bar chart with data points on the bar - Bar charts are very flexible: The height can represent anything, as long as it is a number.
#And each bar can represent anything, as long as it is a category.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
loans = pd.read_csv("prosperLoanData.csv", index_col=0)

ax = loans['EmploymentStatus'].value_counts().plot(kind='barh', figsize=(10,7),
                                        fontsize=13);
ax.set_alpha(0.8)
ax.set_title("Employee status for loan data", fontsize=13, color='blue')
ax.set_xlabel("Count", fontsize=13, color='blue');
ax.set_ylabel("Employee status", fontsize=13, color='blue');


for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=13, color='blue')
# invert for largest on top 
ax.invert_yaxis()
R = mpatches.Patch(color='grey', label='Retired')
NE = mpatches.Patch(color='pink', label='Not Employed')
O = mpatches.Patch(color='violet', label='Others')
P = mpatches.Patch(color='brown', label='Part-time')
NA = mpatches.Patch(color='red', label='Not available')
S = mpatches.Patch(color='green', label='self-employed')
F = mpatches.Patch(color='orange', label='Full-time')
E = mpatches.Patch(color='blue', label='Employed')

ax.legend(handles=[E,F,S,NA,O,P,NE,R], loc='upper right',bbox_to_anchor=(1.25, 1.0))
#ax.legend(loans['EmploymentStatus'], loc='upper right',bbox_to_anchor=(1.35, 1.0),borderaxespad=0)
```

```python
#Line chart - A line chart can pass over any number of many individual values, 
#making it the tool of first choice for distributions with many unique values or categories.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
loans = pd.read_csv("prosperLoanData.csv", index_col=0)
creditlineyear = np.arange(loans['FirstRecordedCreditLine'].shape[0])
yields = loans['EstimatedEffectiveYield']
loss = loans['EstimatedLoss']
returns = loans['EstimatedReturn']

plt.plot(creditlineyear, yields, color='green')
plt.plot(creditlineyear, returns, color='blue')
plt.plot(creditlineyear, loss, color='orange')
plt.xlabel('Estimations', fontsize=13, color='blue')
plt.ylabel('Estimation in millions', fontsize=13, color='blue')
plt.title('Loan Proposal Estimations', fontsize=13, color='blue')
plt.legend(loc='best',bbox_to_anchor=(1, 1))
plt.show()

```

```python
# Area Chart - Area charts are just line charts, but with the bottom shaded in. That's it!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
loans = pd.read_csv("prosperLoanData.csv", index_col=0)
ESD = loans['EmploymentStatusDuration'].sort_index().plot.area()

plt.xlabel('EmploymentStatus', fontsize=13, color='blue')
plt.ylabel('Status in Days', fontsize=13, color='blue')
plt.title('EmploymentStatus Details', fontsize=13, color='blue')
plt.legend(loc='best',bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Histogram - A histogram looks, trivially, like a bar plot. And it basically is! In fact,
#a histogram is special kind of bar plot that splits your data into even intervals and 
#displays how many rows are in each interval with bars. 
#The only analytical difference is that instead of each bar representing a single value, it represents a range of values.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
loans = pd.read_csv("prosperLoanData.csv", index_col=0)
loans[loans['RevolvingCreditBalance'] < 10000]['RevolvingCreditBalance'].plot.hist()

plt.xlabel('CreditBalance',fontsize=13, color='blue')
plt.ylabel('in Rs/-',fontsize=13, color='blue')
plt.title('RevolvingCreditBalance less than 10K',fontsize=13, color='blue')
plt.legend(loc='upper right',bbox_to_anchor=(1, 1))
plt.show()
```

```python
# These slides contains Bivariate charts and their uses in visualization
#Scator Plot - The simplest bivariate plot is the lowly scatter plot.
#A simple scatter plot simply maps each variable of interest to a point in two-dimensional space.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
loans = pd.read_csv("prosperLoanData.csv", index_col=0)
loans[loans['AvailableBankcardCredit'] < 1000].sample(1000).plot.scatter(x='AvailableBankcardCredit', y='TotalInquiries')
plt.xlabel('Card Credit', fontsize=13, color='blue')
plt.ylabel('Inquireies', fontsize=13, color='blue')
plt.title('Bank card Credit details 1K', fontsize=13, color='blue')
plt.show()
```

```python
#Hexplot - The data in this plot is directly comprable to the scatter plot from earlier, 
#but the story it tells us is very different. The hexplot provides us with a much more useful view on the dataset, 
#showing that the bottles of wine reviewed by Wine Magazine cluster around 87.5 points and around $20.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
loans = pd.read_csv("prosperLoanData.csv", index_col=0)
loans[loans['AvailableBankcardCredit'] < 1000].plot.hexbin(x='AvailableBankcardCredit', y='TotalInquiries', gridsize=15)
plt.xlabel('Card Credit', fontsize=13, color='blue')
plt.ylabel('Inquireies', fontsize=13, color='blue')
plt.title('Bank card Credit details 1K', fontsize=13, color='blue')
plt.show()
```

```python
#Stacked plots - Stacked bar plots share the strengths and weaknesses of univariate bar charts. 
#They work best for nominal categorical or small ordinal categorical variables.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cert = pd.read_csv("DACert.csv", index_col=0).plot.bar(stacked=True, figsize=(10,7))
plt.legend(loc='best',bbox_to_anchor=(1.15, 1))
plt.xlabel('Year', fontsize=12, color='blue')
plt.ylabel('Course Details', fontsize=13, color='blue')
plt.title('DataAnalyst course Stats by Year', fontsize=13, color='blue')
plt.show()
```

```python
#Stocked Area plot - Like single-variable area charts, multivariate area charts are meant for nominal categorical or interval variables.
#Stacked plots are visually very pretty

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cert = pd.read_csv("DACert.csv", index_col=0).plot.area(figsize=(10,7))
plt.legend(loc='best',bbox_to_anchor=(1.15, 1))
plt.xlabel('Year', fontsize=12, color='blue')
plt.ylabel('Course Details', fontsize=12, color='blue')
plt.title('DataAnalyst course Stats by Year', fontsize=12, color='blue')
plt.show()
```

```python
#Multivariate plotting - These are multivariate plots used for exploratory visualizations
#scatter plots

#This scatterplot uses three visual variables. The horizontal position (x-value) tracks the Value of the player 
#(how well they are paid). The vertical position (y-value) tracks the Overall score of the player across all attributes. 
#And the color (the hue parameter) tracks which of the three categories of interest the player the point represents is in.

import pandas as pd
pd.set_option('max_columns', None)
df = pd.read_csv("CompleteDataset.csv", index_col=0)

import re
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])

g=sns.lmplot(x='Value', y='Overall', markers=['o', 'x', '*'], hue='Position',
           data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])],
           fit_reg=False
          )
plt.title('Overall Value and Preferred position', fontsize=15, color='blue')
sns.set(font_scale=1.5)
```

```python
#Grouped Box plot - To know answer for the question "doStrikers score higher on "Aggression" than Goalkeepers do?""

f = (footballers
         .loc[footballers['Position'].isin(['ST', 'GK'])]
         .loc[:, ['Value', 'Overall', 'Aggression', 'Position']]
    )
f = f[f["Overall"] >= 80]
f = f[f["Overall"] < 85]
f['Aggression'] = f['Aggression'].astype(float)

g=sns.boxplot(x="Overall", y="Aggression", hue='Position', data=f)
g.legend(loc='best',bbox_to_anchor=(1, 1))
plt.title('Overall Aggression b/n ST and GK', fontsize=15, color='blue')
```

```python
# Heatmap measures the correlation between every pair of values

f = (
    footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']]
        .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan)
        .dropna()
).corr()

sns.heatmap(f, annot=True)
plt.title('Correlation plot for Footballers', fontsize=15, color='blue')
```
